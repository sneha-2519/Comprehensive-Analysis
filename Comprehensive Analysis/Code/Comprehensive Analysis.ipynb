```python
data['wait time '] = pd.to_datetime(data['wait time '])
data['hour'] = data['wait time '].dt.hour
data['minute'] = data['wait time '].dt.minute
data['second'] = data['wait time '].dt.second
data.columns
```

```python
#total minutes = ( (hours * 60) + minutes + (second/60) )
data['Wait minute'] = (data['hour']* 60 + data['minute'] +(data['second']/60))
data['Wait minute'] = round(data['Wait minute'])
data.columns
data.head(5)
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data['day_of_week'] = data['Date'].dt.day_name()
data.head(2)
```

```python
data['Entry Time'] = pd.to_datetime(data['Entry Time'])
data['Entry_hour'] = data['Entry Time'].dt.hour
data.columns
```

```python
wait_time_with_financlial_class = data.groupby('Financial Class')['Wait minute'].sum().reset_index()

# Create a bar chart
plt.bar(wait_time_with_financlial_class['Financial Class'], wait_time_with_financlial_class['Wait minute'])

# Add labels and title to the chart
plt.xlabel("Financial Class")
plt.ylabel("Wait Time (minutes)")
plt.title("Wait Time by Financial Class")

# Show the chart
plt.show()
```
